```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

from exec_boundfit import exec_boundfit
```

```python
infile = 'example.dat'
tabla = np.genfromtxt(infile)
xdata = tabla[:, 0]
ydata = tabla[:, 1]
```

```python
exec_boundfit(infile=infile, fittype=1, poldeg=5, side=1, outbasefilename='example1')
exec_boundfit(infile=infile, fittype=2, poldeg=5, side=1, outbasefilename='example1b')
exec_boundfit(infile=infile, fittype=1, poldeg=5, side=2, outbasefilename='example2')
exec_boundfit(infile=infile, fittype=2, poldeg=5, side=2, outbasefilename='example2b')
exec_boundfit(infile=infile, fittype=3, nknots=15, side=1, nrefine=40, outbasefilename='example3')
exec_boundfit(infile=infile, fittype=3, nknots=15, side=2, nrefine=40, outbasefilename='example4')
```

```python
lfit = np.genfromtxt('example1_lfit.bft')
xpol1 = lfit[:, 0]
ypol1 = lfit[:, 1]
lfit = np.genfromtxt('example1b_lfit.bft')
xpol1b = lfit[:, 0]
ypol1b = lfit[:, 1]
lfit = np.genfromtxt('example2_lfit.bft')
xpol2 = lfit[:, 0]
ypol2 = lfit[:, 1]
lfit = np.genfromtxt('example2b_lfit.bft')
xpol2b = lfit[:, 0]
ypol2b = lfit[:, 1]
lfit = np.genfromtxt('example3_lfit.bft')
xspl1 = lfit[:, 0]
yspl1 = lfit[:, 1]
lfit = np.genfromtxt('example4_lfit.bft')
xspl2 = lfit[:, 0]
yspl2 = lfit[:, 1]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 8))
ax.plot(xdata, ydata, 'ko', markersize=10, label='original data')
ax.plot(xpol1, ypol1, 'b-', label='simple polynomial (generic)')
ax.plot(xpol1b, ypol1b, 'g-', label='simple polynomial (simplified)')
ax.fill_between(xspl1, yspl1, yspl2, facecolor='grey')
ax.plot(xspl1, yspl1, 'r:', label='adaptive splines')
ax.plot(xpol2, ypol2, 'b-')
ax.plot(xpol2b, ypol2b, 'g-')
ax.plot(xspl2, yspl2, 'r:')
ax.set_xlabel('X axis (arbitrary units)')
ax.set_ylabel('Y axis (arbitrary units)')
ax.legend()
```
